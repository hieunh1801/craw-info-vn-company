## Lấy về các mã số thuế của doanh nghiệp trong 1 thành phố

```python
# Example for one Page
import requests
from connection_to_db import execute_count_query
url = "https://thongtindoanhnghiep.co/api/company?l=ha-noi&&r=100&&p=1"
response = requests.get(url)
response.encoding = 'utf-8'
content = response.json()
list_data_sort_company = content["LtsItems"]
total_in_one_city = int(content["Option"]["TotalRow"])
print(total_in_one_city)
for company in list_data_sort_company:
    query = f"""
    INSERT INTO public.company_short(company_tax_number, company_name)
    VALUES ('{company["MaSoThue"]}', '{company['Title']}') 
    ON CONFLICT (company_tax_number) DO NOTHING;"""
    execute_count_query(query)
```

```python
# Example for multiple page in one city
import requests
from connection_to_db import execute_count_query

def generate_data_for_one_city(city_name="ha_noi"):
    page = 1
    while True:
        url = f"https://thongtindoanhnghiep.co/api/company?l={city_name}&&r=100&&p={page}"
        response = requests.get(url)
        response.encoding = 'utf-8'
        content = response.json()
        list_data_sort_company = content["LtsItems"]
        total_in_one_city = int(content["Option"]["TotalRow"]) # tổng số doanh nghiệp trong 1 thành phố
        if page * 100 > total_in_one_city:
            print(page * 100, ' ', total_in_one_city)
            break
        for company in list_data_sort_company:
            query = f"""
            INSERT INTO public.company_short(company_tax_number, company_name)
            VALUES ('{company["MaSoThue"]}', $$escape '{company['Title']}'$$) 
            ON CONFLICT (company_tax_number) DO NOTHING;"""
            execute_count_query(query)
        page = page + 1
        print(page)
generate_data_for_one_city(city_name="ha-noi")
```
